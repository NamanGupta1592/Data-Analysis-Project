### Data Analysis

```python
import pandas as pd

# Load dataset
file_path = '/content/SuperStore_Sales_Dataset.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(df.head())
```

```python
# Shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Basic statistics
print(df.describe())

# Data types
print("Data types of all the columns:::")
print(df.dtypes)

# Missing values
print(df.isnull().sum())
```

```python
# Handle missing values (example: filling missing values with the mean)
# df.fillna(df.mean(), inplace=True)

# Drop columns with a high number of missing values (if any)
df.drop(columns=['ind1'],inplace=True)
df.drop(columns=['ind2'],inplace=True)
df.drop(columns=['Returns'],inplace=True)

# Convert categorical columns to numeric (example: using one-hot encoding)
df = pd.get_dummies(df, drop_first=True)

print(df.head())
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pairplot for a quick visualization
sns.pairplot(df)
plt.show()


# Distribution of target variable (assuming 'Sales' is the target variable)
sns.histplot(df['Sales'], kde=True)
plt.show()

# Box plots for numerical features
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
# Example: creating new features or modifying existing ones
df['Sales_to_Profit_Ratio'] = df['Sales'] / df['Profit']

# Polynomial features
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(df[['Sales', 'Profit']])

# Get feature names (for scikit-learn versions 0.18 and above)
# If you have an older version, you might need to manually define the names
try:
    poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(['Sales', 'Profit']))
except AttributeError:
    poly_df = pd.DataFrame(poly_features, columns=['Sales', 'Profit', 'Sales^2', 'Sales Profit', 'Profit^2'])

df = pd.concat([df, poly_df], axis=1) # Concatenate along columns

print(df.head())
```
